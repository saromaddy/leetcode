```python
from typing import List

nums = [2, 7, 11, 15]
target = 9

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}
        print(seen)
        for i, num in enumerate(nums):
            print(i,num)
            print('sssss',seen)
            complement = target - num
            print(i,complement)
            if complement in seen:
                print('aaaa',complement)
                return [seen[complement], i]
            seen[num] = i
            print(i,i,seen)

# Create object of class
sol = Solution()

# Call function
result = sol.twoSum(nums, target)

# Print output
print(result)
```

```python
from typing import List

class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        print("Input list:", strs)

        if not strs:
            print("Empty list → return ''")
            return ""

        first = strs[0]
        print("Reference word (first):", first)

        for i in range(len(first)):
            print(f"\nChecking index {i}, character '{first[i]}'")

            for word in strs:
                print(f"   Comparing with word: {word}")

                if i >= len(word):
                    print(f"   ❌ Index {i} out of range for '{word}'")
                    print("   Returning:", first[:i])
                    return first[:i]

                if word[i] != first[i]:
                    print(f"   ❌ Mismatch: '{word[i]}' != '{first[i]}'")
                    print("   Returning:", first[:i])
                    return first[:i]

            print(f"   ✅ All words matched for index {i}")

        print("All characters matched → Returning full word:", first)
        return first


# 🔽 Driver code
if __name__ == "__main__":
    sol = Solution()

    print("\n--- Test Case 1 ---")
    strs1 = ["flower","flow","flight"]
    print("Output 1:", sol.longestCommonPrefix(strs1))

    print("\n--- Test Case 2 ---")
    strs2 = ["dog","racecar","car"]
    print("Output 2:", sol.longestCommonPrefix(strs2))

    print("\n--- Test Case 3 ---")
    strs3 = ["apple","app","april"]
    print("Output 3:", sol.longestCommonPrefix(strs3))
```
